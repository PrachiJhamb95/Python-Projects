#Create a Stylized Conceptual Model for Paper 2 - Agricultural Houshehold Model 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Define labor allocated to farming (L_f) from 0 to L_total (conceptual range)
L_f = np.linspace(0, 100, 500)  # 0 to 100 as a proxy for L_total

# Define marginal product functions (qualitative, no specific functional forms)
# MP_h increases with L_f (since L_h = L_total - L_f decreases)
MP_h = 10 + 0.1 * L_f  # Increasing in L_f
# MP_f decreases with L_f under normal conditions
MP_f = 60 - 0.5 * L_f  # Decreasing in L_f
# MP_f under drought (shifted down)
MP_f_drought = 40 - 0.5 * L_f  # Lower than normal
# MP_f under wet shock (shifted up)
MP_f_wet = 80 - 0.5 * L_f  # Higher than normal

# Plot the curves
plt.plot(L_f, MP_h, 'g-', label='MP_h (Hunting)')
plt.plot(L_f, MP_f, 'k-', label='MP_f (Normal)')
plt.plot(L_f, MP_f_drought, 'r--', label='MP_f (Drought)')
plt.plot(L_f, MP_f_wet, 'b--', label='MP_f (Wet Shock)')

# Find approximate intersection points (equilibrium where MP_f = MP_h)
# Normal condition
idx_normal = np.argmin(np.abs(MP_f - MP_h))
L_f_star = L_f[idx_normal]
# Drought
idx_drought = np.argmin(np.abs(MP_f_drought - MP_h))
L_f_drought = L_f[idx_drought]
# Wet shock
idx_wet = np.argmin(np.abs(MP_f_wet - MP_h))
L_f_wet = L_f[idx_wet]

# Mark equilibrium points
plt.axvline(L_f_star, color='k', linestyle=':', alpha=0.5)
plt.axvline(L_f_drought, color='r', linestyle=':', alpha=0.5)
plt.axvline(L_f_wet, color='b', linestyle=':', alpha=0.5)

# Add labels for equilibrium points
plt.text(L_f_drought, max(MP_h) * 0.9, 'L_f_drought', color='r', ha='center')
plt.text(L_f_star, max(MP_h) * 0.95, 'L_f*', color='k', ha='center')
plt.text(L_f_wet, max(MP_h) * 0.9, 'L_f_wet', color='b', ha='center')

# Axes labels
plt.xlabel('Labor Allocated to Farming (L_f)')
plt.ylabel('Marginal Product')
plt.title('Labor Allocation Between Farming and Hunting Under Climate Shocks')

# Add legend
plt.legend()

# Add annotations to explain shifts
plt.annotate('Drought: MP_f falls\n More labor to hunting', 
             xy=(L_f_drought, MP_f_drought[idx_drought]), 
             xytext=(L_f_drought - 20, MP_f_drought[idx_drought] + 20),
             arrowprops=dict(facecolor='red', shrink=0.05), color='red')
plt.annotate('Wet Shock: MP_f rises\n More labor to farming', 
             xy=(L_f_wet, MP_f_wet[idx_wet]), 
             xytext=(L_f_wet + 10, MP_f_wet[idx_wet] + 20),
             arrowprops=dict(facecolor='blue', shrink=0.05), color='blue')

# Grid and display
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig('household_model_figure.png')
```

```python

# Set up the figure
plt.figure(figsize=(10, 6))

# Define labor allocated to farming (L_f) from 0 to 150 to capture all scenarios
L_f = np.linspace(0, 150, 500)

# Define marginal product functions
MP_h = 10 + 0.1 * L_f        # Marginal product of hunting
MP_f = 60 - 0.5 * L_f        # Marginal product of farming (normal)
MP_f_drought = 40 - 0.5 * L_f  # Marginal product of farming (drought)
MP_f_wet = 80 - 0.5 * L_f    # Marginal product of farming (wet shock)

# Plot the curves
plt.plot(L_f, MP_h, 'g-', label='MP_h (Hunting)')
plt.plot(L_f, MP_f, 'k-', label='MP_f (Normal)')
plt.plot(L_f, MP_f_drought, 'r--', label='MP_f (Drought)')
plt.plot(L_f, MP_f_wet, 'b--', label='MP_f (Wet Shock)')

# Find approximate intersection points (equilibrium where MP_f = MP_h)
idx_normal = np.argmin(np.abs(MP_f - MP_h))
L_f_star = L_f[idx_normal]
idx_drought = np.argmin(np.abs(MP_f_drought - MP_h))
L_f_drought = L_f[idx_drought]
idx_wet = np.argmin(np.abs(MP_f_wet - MP_h))
L_f_wet = L_f[idx_wet]

# Plot vertical lines at equilibrium labor allocations
plt.axvline(L_f_star, color='k', linestyle=':', alpha=0.5)
plt.axvline(L_f_drought, color='r', linestyle=':', alpha=0.5)
plt.axvline(L_f_wet, color='b', linestyle=':', alpha=0.5)

# Add markers at intersection points for clarity
plt.plot(L_f_star, MP_f[idx_normal], 'ko', markersize=8)
plt.plot(L_f_drought, MP_f_drought[idx_drought], 'ro', markersize=8)
plt.plot(L_f_wet, MP_f_wet[idx_wet], 'bo', markersize=8)

# Add labels for equilibrium points
plt.text(L_f_drought, max(MP_h) * 0.9, 'L_f_drought', color='r', ha='center')
plt.text(L_f_star, max(MP_h) * 0.95, 'L_f*', color='k', ha='center')
plt.text(L_f_wet, max(MP_h) * 0.9, 'L_f_wet', color='b', ha='right')

# Axes labels and title
plt.xlabel('Labor Allocated to Farming (L_f)')
plt.ylabel('Marginal Product')
plt.title('Labor Allocation Between Farming and Hunting Under Climate Shocks')

# Add legend
plt.legend()

# Add annotations to explain shifts
plt.annotate('Drought: MP_f falls\n- More labor to hunting', 
             xy=(L_f_drought, MP_f_drought[idx_drought]), 
             xytext=(L_f_drought - 40, MP_f_drought[idx_drought] + 20))
             #arrowprops=dict(facecolor='red', shrink=0.02), color='red')
plt.annotate('Wet Shock: MP_f rises\n- More labor to farming', 
             xy=(L_f_wet, MP_f_wet[idx_wet]), 
             xytext=(L_f_wet + 2, MP_f_wet[idx_wet] + 4))
             #arrowprops=dict(facecolor='blue', shrink=0.02), color='blue')

# Add text box to explain equilibrium condition
#plt.text(120, 70, "Equilibrium where MP_f = MP_h", fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

# Add grid and set y-axis limit
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(0, 85)

# Save the figure
plt.savefig('household_model_figure_updated.png')
```
